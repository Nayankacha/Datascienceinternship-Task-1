**STEP 2 = IMPORT THE LIBRARIES** 

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io 
from PIL import Image 
import matplotlib.pylab as plt
```

**STEP 3 = WE WILL READ THE IMAGE FROM URLS AND DISPLAY THEM USING OPENCV**

```python
image=io.imread("ganpati-wallpapers-20.jpg")
cv2_imshow(image)
```

**STEP 4 = WE WILL PERFORM THE OPERATION ON THE IMAGE** 

```python
grey_filter = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
cv2_imshow(grey_filter)
  # grey scale image generate a real image
```

```python
invert = cv.bitwise_not(grey_filter)
cv2_imshow(invert)
```

```python
blur=cv.GaussianBlur(invert,(21,21),0)
cv2_imshow(blur)
```

```python
invertedblur=cv.bitwise_not(blur)
cv2_imshow(invertedblur)
```

**STEP 5 = AFTER PERFORMING OPERATION WE WILL GET THE PENCIL SKETCHING IMAGE**

```python
sketch_filter=cv.divide(grey_filter,invertedblur,scale=255.0)
cv2_imshow(sketch_filter)
cv.imwrite("sketch.jpg",sketch_filter)
```
